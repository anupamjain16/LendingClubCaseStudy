**Univariate & Bivariate analysis:**
Implemented Univariate analysis on Home Ownership.
Implemented Bivariate analysis to know impact of home ownership on funded amount.



```python
#By Neelam
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns

# Assuming you have a DataFrame named 'loan'
loan = pd.read_csv('LendingClubCaseStudy - loan.csv')

# Create subsets based on 'home_ownership' categories
rent_home = loan[loan['home_ownership'] == 'RENT']
own_home = loan[loan['home_ownership'] == 'OWN']
mortgage_home = loan[loan['home_ownership'] == 'MORTGAGE']

# Count the number of occurrences for each category
counts = [len(rent_home), len(own_home), len(mortgage_home)]

# Labels for the pie chart
labels = ['RENT', 'OWN', 'MORTGAGE']
#univariant analysis
# Plot a pie chart
plt.figure(figsize=(8, 8))
plt.pie(counts, labels=labels, autopct='%1.1f%%', startangle=140, colors=['lightcoral', 'lightblue', 'lightgreen'])
plt.title('Distribution of Home Ownership')
plt.show()

#bivariate Analysis
# Analyze the average Funded amount for each group
avg_rent = rent_home['funded_amnt_inv'].mean()
avg_own = own_home['funded_amnt_inv'].mean()
avg_mortgage = mortgage_home['funded_amnt_inv'].mean()


# Visualize the results with a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=['RENT', 'OWN', 'MORTGAGE'], y=[avg_rent, avg_own, avg_mortgage])
plt.title('Impact of home ownership on funded amount')
plt.xlabel('home_ownership ')
plt.ylabel('Average funded amount')
plt.show()
```

**Bivariate analysis**:  Implemented Bivariate analysis to know impact of purpose on loan_status.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
loan = pd.read_csv('LendingClubCaseStudy - loan.csv', error_bad_lines=False)

# Convert loan_status to categorical for proper ordering in plots
loan['loan_status'] = pd.Categorical(loan['loan_status'], categories=['Current', 'Fully Paid', 'Charged Off'], ordered=True)

# Count the occurrences of each purpose for each loan status
purpose_status_counts = loan.groupby(['purpose', 'loan_status']).size().unstack(fill_value=0)

# Plotting the bar chart
plt.figure(figsize=(18, 6))
purpose_status_counts.plot(kind='bar', stacked=True)
plt.title('Loan Status Distribution for Each Purpose')
plt.xlabel('Purpose')
plt.ylabel('Count')
plt.xticks(rotation=75)
plt.legend(title='Loan Status')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'LendingClubCaseStudy - loan.csv'  # Replace with your file path
loan_data = pd.read_csv(file_path)

# Univariate Analysis
plt.figure(figsize=(18, 6))

# Plot for grade
sns.countplot(x='grade', data=loan_data, order=sorted(loan_data['grade'].unique()))
plt.title('Loan Grade Distribution')
plt.xlabel('Grade')
plt.ylabel('Count')

plt.tight_layout()
plt.show()

# Bivariate Analysis
plt.figure(figsize=(18, 6))

# Box plot for loan_amnt vs grade
sns.boxplot(x='grade', y='funded_amnt_inv', data=loan_data, order=sorted(loan_data['grade'].unique()))
plt.title('Funded Amount by Grade')
plt.xlabel('Grade')
plt.ylabel('Funded Amount')

plt.tight_layout()
plt.show()
```

```python
#By Anupam Jain
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the loan data from CSV file
loan_data = pd.read_csv('LendingClubCaseStudy - loan.csv')

# Display basic information about the dataset
print(loan_data.info())

# Display summary statistics
print(loan_data.describe())

# Check for missing values
print(loan_data.isnull().sum())

# Distribution of loan status
sns.countplot(x='loan_status', data=loan_data)
plt.title('Distribution of Loan Status')
plt.show()

# Distribution of loan amounts
plt.figure(figsize=(12, 6))
sns.histplot(loan_data['loan_amnt'], bins=30, kde=False)
plt.title('Distribution of Loan Amounts')
plt.xlabel('Loan Amount')
plt.ylabel('Count')
plt.show()

# Loan status based on purpose
plt.figure(figsize=(14, 7))
sns.countplot(x='purpose', hue='loan_status', data=loan_data)
plt.title('Loan Status Based on Purpose')
plt.xticks(rotation=45, ha='right')
plt.show()

# Loan status based on grade
sns.countplot(x='grade', hue='loan_status', data=loan_data, order=sorted(loan_data['grade'].unique()))
plt.title('Loan Status Based on Grade')
plt.show()

# Correlation matrix
correlation_matrix = loan_data.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()

# Pair plot for selected columns
sns.pairplot(loan_data[['loan_amnt', 'installment', 'annual_inc', 'dti', 'total_pymnt', 'loan_status']], hue='loan_status')
plt.suptitle('Pair Plot for Selected Columns')
plt.show()

# Box plot for interest rate based on loan status
plt.figure(figsize=(12, 6))
#sns.boxplot(x='loan_status', y='int_rate', data=loan_data)
plt.title('Interest Rate Distribution Based on Loan Status')
plt.show()

# Employment length vs Loan status
plt.figure(figsize=(14, 7))
#sns.countplot(x='emp_length', hue='loan_status', data=loan_data, order=sorted(loan_data['emp_length'].unique()))
plt.title('Loan Status Based on Employment Length')
plt.show()


# Employment length vs Loan status
plt.figure(figsize=(14, 7))
sns.countplot(x='emp_length', hue='loan_status', data=loan_data, order=sorted(loan_data['emp_length'].unique()))
plt.title('Loan Status Based on Employment Length')
plt.show()
```
